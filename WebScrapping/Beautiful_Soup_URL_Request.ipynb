## SUMMING UP ALL WE DID UNTIL NOW and Generating CSV File

```python
import csv
```

```python
csv_file=open('coreyms_scrape.csv','w') # we want to write to this file
```

```python
csv_writer=csv.writer(csv_file)
```

```python
csv_writer.writerow(['headline','summary','youtube-link']) # Column names
```

```python
for article in soup.find_all('article'):
    headline=article.header.h2.a.text
    print(headline)
    content=article.find('div')
    print(content.p.text)
    y_link=None
    try:
        video_source=content.find('iframe')['src']
        video_id=video_source.split('/')[4].split('?')[0]
        y_link='https://youtube.com/watch?v='+video_id
    except Exception as e:
        y_link=None
    print(y_link)
    csv_writer.writerow([headline,content.p.text,y_link])
    print()
```

```python
csv_file.close()
```
