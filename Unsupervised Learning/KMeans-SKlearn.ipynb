```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[1,2],[1.5,1.8],[5,8],[8,8],[1,0.6],[9,11]])
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters=2) # By default the clusters are 8
```

```python
k_means.fit(x)
# init = 'k-means++' means that the initially the means of the cluster that are chosen are spread out and not close to each other
# n_init = 10 means that we will choose the intial k mean points the 1st time and whichever initialization gives us the best result , we will use that and the process of choosing the best k mean values initially will be done maximum of 10 times 
# So, whichever initialization gives us the best result in terms of minimizing the cost function , we will pick that initialization
# max_iter is the number of times we want update our clusters and mean
```

```python
k_means.labels_
```

```python
k_means.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=k_means.labels_) # different color for each cluster 
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1])
```
